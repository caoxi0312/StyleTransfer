```python
df.NUMBER_OF_EMPLOYEES.describe()
```

```python
y = df.NUMBER_OF_EMPLOYEES.copy()
for com in y:
    if com < 10:
        y.replace(com, '1-9', inplace=True)
    elif com < 20:
        y.replace(com, '10-19', inplace=True)
    elif com < 100:
        y.replace(com, '20-99', inplace=True)
    elif com < 500:
        y.replace(com, '100-499', inplace=True)
    elif com < 1000:
        y.replace(com, '500-999', inplace=True)
    else:
        y.replace(com, '>1000', inplace=True)
```

```python
size = y.groupby(y).count()
size
```

```python
fig = plt.figure(figsize=(10,4))
plt.hist(df.NUMBER_OF_EMPLOYEES, bins=20,range=(1,200))
plt.xlabel("NUMBER OF EMPLOYEES")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("size.png",dpi=500)
```
